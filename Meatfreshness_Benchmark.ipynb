```python
import os, math, random, hashlib, warnings
from dataclasses import dataclass, asdict
from pathlib import Path
from typing import List, Dict, Tuple, Optional

import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.optim as optim

from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms as T

from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True  # avoid crashes on truncated files

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

import matplotlib.pyplot as plt
from tqdm import tqdm

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("DEVICE:", DEVICE)
warnings.filterwarnings("once", category=UserWarning)


# =========================================================
# 1) Config
# =========================================================
@dataclass
class Config:
    ROOT: str = "./Meat Freshness.v1-new-dataset.multiclass"  # <-- CHANGE THIS

    # Splits
    TEST_SIZE: float = 0.30          # total test split from all data
    VAL_FROM_TRAIN: float = 0.20     # val split from train portion

    # DataLoader
    BATCH_SIZE: int = 16
    NUM_WORKERS: int = 4

    # Training
    EPOCHS: int = 30
    LR: float = 3e-4
    WEIGHT_DECAY: float = 1e-4
    LABEL_SMOOTHING: float = 0.0     # try 0.05–0.1 if overfitting
    FREEZE_BACKBONE_EPOCHS: int = 0  # e.g. 2–5 for tiny datasets

    # Early stopping
    PATIENCE: int = 6
    MIN_DELTA: float = 0.0

    # Stability
    USE_AMP: bool = True
    GRAD_CLIP_NORM: float = 0.0      # e.g. 1.0 to enable
    BASE_SEED: int = 2023
    DETERMINISTIC: bool = False

    # Imbalance handling
    USE_WEIGHTED_SAMPLER: bool = True
    USE_WEIGHTED_LOSS: bool = False

    # Augmentation policy: "none" | "light" | "strong" | "randaugment"
    AUGMENTATION: str = "strong"

    # Saving
    SAVE_DIR: str = "saved_models"
    RESULTS_DIR: str = "results"
    SAVE_PREFIX: str = "meat"

    # Model zoo to benchmark
    MODELS_TO_RUN: Tuple[str, ...] = (
        "resnet50",
        "efficientnet_b0",
        "mobilenet_v3_large",
        "densenet121",
        "vit_b_16",
        "efficientnet_v2_s",  # if available
        "convnext_tiny",      # if available
        "swin_t",             # if available
    )

cfg = Config()
print(pd.Series(asdict(cfg)))


# =========================================================
# 2) Reproducibility
# =========================================================
def set_seed(seed: int, deterministic: bool = False):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    if deterministic:
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    else:
        torch.backends.cudnn.benchmark = True

def seed_worker(worker_id: int):
    ws = torch.initial_seed() % 2**32
    np.random.seed(ws)
    random.seed(ws)

set_seed(cfg.BASE_SEED, cfg.DETERMINISTIC)


# =========================================================
# 3) Data discovery + corruption filter + leakage checks
# =========================================================
IMG_EXTS = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

def resolve_image_dir(base: Path) -> Path:
    # dataset sometimes has train/images
    return base / "images" if (base / "images").exists() else base

def list_images(folder: Path) -> List[Path]:
    files = []
    for ext in IMG_EXTS:
        files.extend(folder.rglob(f"*{ext}"))
        files.extend(folder.rglob(f"*{ext.upper()}"))
    return sorted(set(files))

def infer_class_from_path(p: Path) -> str:
    # Primary: prefix before '-' in filename: FRESH-001.jpg -> FRESH
    if "-" in p.name:
        return p.name.split("-")[0].strip()
    # Fallback: parent folder name (if class-folder structure)
    return p.parent.name.strip()

def is_image_ok(path: str) -> bool:
    # Verify image can be opened & decoded
    try:
        with Image.open(path) as im:
            im.verify()
        return True
    except Exception:
        return False

def load_samples(root: str):
    root = Path(root)
    train_dir = resolve_image_dir(root / "train")
    valid_dir = resolve_image_dir(root / "valid")

    if not train_dir.exists() or not valid_dir.exists():
        raise FileNotFoundError(
            f"Expected train/ and valid/ under {root}\n"
            f"train exists={train_dir.exists()} | valid exists={valid_dir.exists()}"
        )

    all_paths = list_images(train_dir) + list_images(valid_dir)
    if len(all_paths) == 0:
        raise FileNotFoundError(f"No images found under {train_dir} or {valid_dir}.")

    # filter corrupt images (important for “professional edge-case handling”)
    ok_paths = []
    bad = 0
    for p in tqdm(all_paths, desc="scan images"):
        if is_image_ok(str(p)):
            ok_paths.append(p)
        else:
            bad += 1
    if bad > 0:
        print(f"[WARN] Dropped {bad} corrupt/unreadable images.")

    class_names = sorted(list({infer_class_from_path(p) for p in ok_paths}))
    class_to_idx = {c: i for i, c in enumerate(class_names)}

    samples = [(str(p), class_to_idx[infer_class_from_path(p)]) for p in ok_paths]
    return samples, class_to_idx

samples, classes = load_samples(cfg.ROOT)
idx_to_class = {v: k for k, v in classes.items()}
num_classes = len(classes)

print("Total usable images:", len(samples))
print("Classes:", classes)

# distribution
dist = {}
for _, y in samples:
    dist[y] = dist.get(y, 0) + 1
print("Distribution:", {idx_to_class[k]: v for k, v in dist.items()})

X_paths = np.array([p for p, _ in samples])
y_labels = np.array([y for _, y in samples])

def safe_stratified_split(X, y, test_size, seed, name="split"):
    try:
        return train_test_split(X, y, test_size=test_size, random_state=seed, stratify=y)
    except ValueError as e:
        print(f"[WARN] Stratified {name} failed: {e}")
        print("[WARN] Falling back to non-stratified split.")
        return train_test_split(X, y, test_size=test_size, random_state=seed, stratify=None)

# train/test, then train/val
X_train, X_test, y_train, y_test = safe_stratified_split(
    X_paths, y_labels, cfg.TEST_SIZE, cfg.BASE_SEED, "train/test"
)
X_train, X_val, y_train, y_val = safe_stratified_split(
    X_train, y_train, cfg.VAL_FROM_TRAIN, cfg.BASE_SEED, "train/val"
)

print("Split sizes:", "train", len(X_train), "val", len(X_val), "test", len(X_test))

def md5_file(path: str, chunk_size: int = 1 << 20) -> str:
    h = hashlib.md5()
    with open(path, "rb") as f:
        while True:
            b = f.read(chunk_size)
            if not b:
                break
            h.update(b)
    return h.hexdigest()

def overlap_hash_count(a_paths, b_paths):
    a = {md5_file(p) for p in a_paths}
    b = {md5_file(p) for p in b_paths}
    return len(a & b)

print("Exact duplicate overlaps (MD5):")
print("train ∩ val :", overlap_hash_count(X_train, X_val))
print("train ∩ test:", overlap_hash_count(X_train, X_test))
print("val ∩ test  :", overlap_hash_count(X_val, X_test))


# =========================================================
# 4) Augmentations (small dataset options)
# =========================================================
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]

def get_input_size(model_name: str) -> int:
    # ViT expects 224 safely. CNNs can take larger.
    return 224 if model_name.lower() == "vit_b_16" else 416

def build_transforms(input_size: int, policy: str, train: bool) -> T.Compose:
    if not train:
        return T.Compose([
            T.Resize((input_size, input_size)),
            T.ToTensor(),
            T.Normalize(IMAGENET_MEAN, IMAGENET_STD),
        ])

    policy = policy.lower()
    aug = []

    if policy == "none":
        aug = [T.Resize((input_size, input_size))]

    elif policy == "light":
        aug = [
            T.RandomResizedCrop(input_size, scale=(0.85, 1.0), ratio=(0.9, 1.1)),
            T.RandomHorizontalFlip(0.5),
            T.RandomRotation(8),
            T.ColorJitter(brightness=0.10, contrast=0.10, saturation=0.05, hue=0.02),
            T.RandomAdjustSharpness(sharpness_factor=1.5, p=0.2),
        ]

    elif policy == "strong":
        aug = [
            T.RandomResizedCrop(input_size, scale=(0.70, 1.0), ratio=(0.85, 1.15)),
            T.RandomHorizontalFlip(0.5),
            T.RandomRotation(15),
            T.RandomAffine(degrees=0, translate=(0.05, 0.05), scale=(0.95, 1.05), shear=5),
            T.ColorJitter(brightness=0.20, contrast=0.20, saturation=0.15, hue=0.04),
            T.RandomAutocontrast(p=0.2),
            T.RandomEqualize(p=0.1),
        ]
        if hasattr(T, "GaussianBlur"):
            aug.append(T.GaussianBlur(kernel_size=3, sigma=(0.1, 1.5)))

    elif policy == "randaugment":
        aug = [
            T.RandomResizedCrop(input_size, scale=(0.70, 1.0), ratio=(0.85, 1.15)),
            T.RandomHorizontalFlip(0.5),
        ]
        if hasattr(T, "RandAugment"):
            aug.append(T.RandAugment(num_ops=2, magnitude=9))
        else:
            print("[WARN] RandAugment not available; using strong policy fallback.")
            aug += [
                T.RandomRotation(15),
                T.ColorJitter(0.20, 0.20, 0.15, 0.04),
            ]
    else:
        raise ValueError(f"Unknown AUGMENTATION policy: {policy}")

    aug += [
        T.ToTensor(),
        T.Normalize(IMAGENET_MEAN, IMAGENET_STD),
    ]

    if hasattr(T, "RandomErasing"):
        aug.append(T.RandomErasing(p=0.25, scale=(0.02, 0.12), ratio=(0.3, 3.3), value="random"))

    return T.Compose(aug)


# =========================================================
# 5) Dataset + Loaders (class balancing supported)
# =========================================================
class MeatImageDataset(Dataset):
    def __init__(self, paths, labels, transform=None):
        self.paths = list(paths)
        self.labels = list(labels)
        self.transform = transform

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        p = self.paths[idx]
        y = int(self.labels[idx])
        img = Image.open(p).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, y

def make_loaders(model_name: str):
    input_size = get_input_size(model_name)

    tfm_train = build_transforms(input_size, cfg.AUGMENTATION, train=True)
    tfm_eval  = build_transforms(input_size, "none", train=False)

    train_ds = MeatImageDataset(X_train, y_train, tfm_train)
    val_ds   = MeatImageDataset(X_val, y_val, tfm_eval)
    test_ds  = MeatImageDataset(X_test, y_test, tfm_eval)

    sampler = None
    if cfg.USE_WEIGHTED_SAMPLER:
        counts = np.bincount(y_train, minlength=num_classes).astype(np.float64)
        counts = np.maximum(counts, 1.0)
        class_w = 1.0 / counts
        sample_w = class_w[y_train]
        sampler = WeightedRandomSampler(
            weights=torch.as_tensor(sample_w, dtype=torch.double),
            num_samples=len(sample_w),
            replacement=True
        )

    g = torch.Generator()
    g.manual_seed(cfg.BASE_SEED)

    dl_kwargs = dict(
        num_workers=cfg.NUM_WORKERS,
        pin_memory=(DEVICE.type == "cuda"),
        worker_init_fn=seed_worker,
        generator=g
    )
    if cfg.NUM_WORKERS > 0:
        dl_kwargs["persistent_workers"] = True
        dl_kwargs["prefetch_factor"] = 2

    train_dl = DataLoader(
        train_ds, batch_size=cfg.BATCH_SIZE,
        shuffle=(sampler is None), sampler=sampler, drop_last=False, **dl_kwargs
    )
    val_dl = DataLoader(
        val_ds, batch_size=cfg.BATCH_SIZE,
        shuffle=False, drop_last=False, **dl_kwargs
    )
    test_dl = DataLoader(
        test_ds, batch_size=1,
        shuffle=False, drop_last=False, **dl_kwargs
    )
    return train_dl, val_dl, test_dl


# =========================================================
# 6) Model factory (modern torchvision weights API)
# =========================================================
def build_model(model_name: str, num_classes: int, pretrained: bool = True) -> nn.Module:
    name = model_name.lower()

    # Import lazily so notebook runs even if some are missing
    import torchvision.models as M

    def _try(getter, weights_enum_name: Optional[str], replace_head_fn):
        if not hasattr(M, getter):
            raise AttributeError(f"{getter} not in torchvision.models")
        fn = getattr(M, getter)
        weights = None
        if pretrained and weights_enum_name and hasattr(M, weights_enum_name):
            weights = getattr(M, weights_enum_name).DEFAULT
            model = fn(weights=weights)
        else:
            # fallback (older torchvision)
            try:
                model = fn(pretrained=pretrained)
            except TypeError:
                model = fn()
        replace_head_fn(model)
        return model

    if name == "resnet50":
        return _try("resnet50", "ResNet50_Weights",
                    lambda m: setattr(m, "fc", nn.Linear(m.fc.in_features, num_classes)))

    if name == "densenet121":
        return _try("densenet121", "DenseNet121_Weights",
                    lambda m: setattr(m, "classifier", nn.Linear(m.classifier.in_features, num_classes)))

    if name == "efficientnet_b0":
        return _try("efficientnet_b0", "EfficientNet_B0_Weights",
                    lambda m: setattr(m.classifier, "1", nn.Linear(m.classifier[1].in_features, num_classes)))

    if name == "efficientnet_v2_s":
        return _try("efficientnet_v2_s", "EfficientNet_V2_S_Weights",
                    lambda m: setattr(m.classifier, "1", nn.Linear(m.classifier[1].in_features, num_classes)))

    if name == "mobilenet_v3_large":
        return _try("mobilenet_v3_large", "MobileNet_V3_Large_Weights",
                    lambda m: setattr(m.classifier, "3", nn.Linear(m.classifier[3].in_features, num_classes)))

    if name == "convnext_tiny":
        return _try("convnext_tiny", "ConvNeXt_Tiny_Weights",
                    lambda m: setattr(m.classifier, "2", nn.Linear(m.classifier[2].in_features, num_classes)))

    if name == "swin_t":
        return _try("swin_t", "Swin_T_Weights",
                    lambda m: setattr(m.head, "fc", nn.Linear(m.head.in_features, num_classes)) if hasattr(m.head,"fc")
                    else setattr(m, "head", nn.Linear(m.head.in_features, num_classes)))

    if name == "vit_b_16":
        def repl(m):
            # torchvision ViT uses m.heads.head
            if hasattr(m, "heads") and hasattr(m.heads, "head"):
                m.heads.head = nn.Linear(m.heads.head.in_features, num_classes)
            elif hasattr(m, "head"):
                m.head = nn.Linear(m.head.in_features, num_classes)
            else:
                raise AttributeError("Unknown ViT head structure.")
        return _try("vit_b_16", "ViT_B_16_Weights", repl)

    raise ValueError(f"Unknown model_name: {model_name}")


# =========================================================
# 7) Training + Evaluation
# =========================================================
def compute_class_weights(y: np.ndarray, num_classes: int) -> torch.Tensor:
    counts = np.bincount(y, minlength=num_classes).astype(np.float32)
    counts = np.maximum(counts, 1.0)
    w = counts.sum() / (num_classes * counts)
    return torch.tensor(w, dtype=torch.float32, device=DEVICE)

def make_loss_fn():
    if cfg.USE_WEIGHTED_LOSS:
        w = compute_class_weights(y_train, num_classes)
        return nn.CrossEntropyLoss(weight=w, label_smoothing=cfg.LABEL_SMOOTHING)
    return nn.CrossEntropyLoss(label_smoothing=cfg.LABEL_SMOOTHING)

def make_optimizer(model: nn.Module):
    return optim.AdamW(model.parameters(), lr=cfg.LR, weight_decay=cfg.WEIGHT_DECAY)

def to_device(batch):
    ims, gts = batch
    ims = ims.to(DEVICE, non_blocking=True)
    gts = torch.as_tensor(gts, device=DEVICE).long()
    return ims, gts

def freeze_backbone(model_name: str, model: nn.Module, freeze: bool):
    # Freeze everything except classifier head
    # (model-specific heads handled below)
    for p in model.parameters():
        p.requires_grad = (not freeze)

    name = model_name.lower()
    if name == "resnet50":
        for p in model.fc.parameters(): p.requires_grad = True
    elif name == "densenet121":
        for p in model.classifier.parameters(): p.requires_grad = True
    elif "efficientnet" in name:
        for p in model.classifier.parameters(): p.requires_grad = True
    elif name == "mobilenet_v3_large":
        for p in model.classifier.parameters(): p.requires_grad = True
    elif name == "convnext_tiny":
        for p in model.classifier.parameters(): p.requires_grad = True
    elif name == "vit_b_16":
        if hasattr(model, "heads"):
            for p in model.heads.parameters(): p.requires_grad = True
        elif hasattr(model, "head"):
            for p in model.head.parameters(): p.requires_grad = True

@torch.no_grad()
def evaluate(model, loader, loss_fn):
    model.eval()
    total_loss = 0.0
    correct = 0
    for ims, gts in tqdm(loader, desc="val", leave=False):
        ims, gts = to_device((ims, gts))
        logits = model(ims)
        loss = loss_fn(logits, gts)
        total_loss += loss.item()
        correct += (logits.argmax(1) == gts).sum().item()
    return total_loss / max(1, len(loader)), correct / len(loader.dataset)

def train_one_epoch(model, loader, loss_fn, optimizer, scaler):
    model.train()
    total_loss = 0.0
    correct = 0
    use_amp = (cfg.USE_AMP and DEVICE.type == "cuda")

    for ims, gts in tqdm(loader, desc="train", leave=False):
        ims, gts = to_device((ims, gts))
        optimizer.zero_grad(set_to_none=True)

        if use_amp:
            with torch.cuda.amp.autocast():
                logits = model(ims)
                loss = loss_fn(logits, gts)
            scaler.scale(loss).backward()
            if cfg.GRAD_CLIP_NORM and cfg.GRAD_CLIP_NORM > 0:
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), cfg.GRAD_CLIP_NORM)
            scaler.step(optimizer)
            scaler.update()
        else:
            logits = model(ims)
            loss = loss_fn(logits, gts)
            loss.backward()
            if cfg.GRAD_CLIP_NORM and cfg.GRAD_CLIP_NORM > 0:
                torch.nn.utils.clip_grad_norm_(model.parameters(), cfg.GRAD_CLIP_NORM)
            optimizer.step()

        total_loss += loss.item()
        correct += (logits.argmax(1) == gts).sum().item()

    return total_loss / max(1, len(loader)), correct / len(loader.dataset)

def fit(model_name: str, seed: int):
    set_seed(seed, cfg.DETERMINISTIC)

    train_dl, val_dl, test_dl = make_loaders(model_name)
    model = build_model(model_name, num_classes, pretrained=True).to(DEVICE)

    # Optional: freeze backbone first
    if cfg.FREEZE_BACKBONE_EPOCHS > 0:
        freeze_backbone(model_name, model, freeze=True)

    loss_fn = make_loss_fn()
    optimizer = make_optimizer(model)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode="min", factor=0.5, patience=2)

    scaler = torch.cuda.amp.GradScaler(enabled=(cfg.USE_AMP and DEVICE.type == "cuda"))

    save_dir = Path(cfg.SAVE_DIR) / model_name
    save_dir.mkdir(parents=True, exist_ok=True)
    ckpt_path = save_dir / f"{cfg.SAVE_PREFIX}_{model_name}_seed{seed}_best.pth"

    best_loss = float("inf")
    no_improve = 0
    history = []

    for epoch in range(cfg.EPOCHS):
        # unfreeze after N epochs
        if cfg.FREEZE_BACKBONE_EPOCHS > 0 and epoch == cfg.FREEZE_BACKBONE_EPOCHS:
            freeze_backbone(model_name, model, freeze=False)

        tr_loss, tr_acc = train_one_epoch(model, train_dl, loss_fn, optimizer, scaler)
        va_loss, va_acc = evaluate(model, val_dl, loss_fn)
        scheduler.step(va_loss)

        history.append({
            "epoch": epoch + 1,
            "train_loss": tr_loss, "train_acc": tr_acc,
            "val_loss": va_loss, "val_acc": va_acc,
            "lr": optimizer.param_groups[0]["lr"]
        })

        print(f"[{model_name}] Epoch {epoch+1:02d} | "
              f"tr_loss={tr_loss:.4f} tr_acc={tr_acc:.4f} | "
              f"va_loss={va_loss:.4f} va_acc={va_acc:.4f} | lr={optimizer.param_groups[0]['lr']:.2e}")

        if va_loss < (best_loss - cfg.MIN_DELTA):
            best_loss = va_loss
            no_improve = 0
            torch.save(model.state_dict(), ckpt_path)
            print("  Saved best checkpoint.")
        else:
            no_improve += 1
            if no_improve >= cfg.PATIENCE:
                print(f"  Early stop (no improvement for {cfg.PATIENCE} epochs).")
                break

    return str(ckpt_path), pd.DataFrame(history), test_dl

@torch.no_grad()
def predict(model, loader):
    model.eval()
    y_true, y_pred = [], []
    for ims, gts in tqdm(loader, desc="test", leave=False):
        ims = ims.to(DEVICE, non_blocking=True)
        gts = torch.as_tensor(gts, device=DEVICE).long()
        logits = model(ims)
        pred = logits.argmax(1)
        y_true.append(int(gts.item()))
        y_pred.append(int(pred.item()))
    return np.array(y_true), np.array(y_pred)

def macro_sensitivity_specificity(cm: np.ndarray):
    K = cm.shape[0]
    total = cm.sum()
    sens, spec = [], []
    for c in range(K):
        TP = cm[c, c]
        FN = cm[c, :].sum() - TP
        FP = cm[:, c].sum() - TP
        TN = total - (TP + FN + FP)
        sens.append(TP / (TP + FN + 1e-12))
        spec.append(TN / (TN + FP + 1e-12))
    return float(np.mean(sens)), float(np.mean(spec)), sens, spec

def plot_confusion(cm: np.ndarray, class_names: List[str], title="Confusion Matrix"):
    plt.figure(figsize=(6,5))
    plt.imshow(cm, interpolation="nearest")
    plt.title(title)
    plt.colorbar()
    ticks = np.arange(len(class_names))
    plt.xticks(ticks, class_names, rotation=45, ha="right")
    plt.yticks(ticks, class_names)

    thresh = cm.max() * 0.5
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, str(cm[i, j]), ha="center", va="center",
                     color="white" if cm[i, j] > thresh else "black")
    plt.ylabel("True")
    plt.xlabel("Pred")
    plt.tight_layout()
    plt.show()


# =========================================================
# 8) Runner: single model or benchmark
# =========================================================
target_names = [idx_to_class[i] for i in range(num_classes)]
print("target_names:", target_names)

def run_single(model_name="resnet50", seed=None):
    seed = cfg.BASE_SEED if seed is None else seed
    ckpt, hist, test_dl = fit(model_name, seed)

    model = build_model(model_name, num_classes, pretrained=False).to(DEVICE)
    model.load_state_dict(torch.load(ckpt, map_location=DEVICE))

    y_true, y_pred = predict(model, test_dl)
    cm = confusion_matrix(y_true, y_pred)

    print("Confusion Matrix (rows=true, cols=pred):\n", cm)
    print("\nClassification Report:\n")
    print(classification_report(y_true, y_pred, target_names=target_names, digits=4))

    macro_sens, macro_spec, sens_list, spec_list = macro_sensitivity_specificity(cm)
    print(f"Macro Sensitivity (Recall): {macro_sens:.4f}")
    print(f"Macro Specificity:          {macro_spec:.4f}")

    plot_confusion(cm, target_names, title=f"{model_name} | seed={seed}")

    per_class = pd.DataFrame({
        "class": target_names,
        "sensitivity(recall)": sens_list,
        "specificity": spec_list
    })
    return ckpt, hist, per_class

def run_benchmark(models: List[str], seeds: List[int]):
    os.makedirs(cfg.RESULTS_DIR, exist_ok=True)
    rows = []

    for m in models:
        # Skip models that torchvision build doesn't have
        try:
            _ = build_model(m, num_classes, pretrained=False)
        except Exception as e:
            print(f"[SKIP] {m} not available in your torchvision build: {e}")
            continue

        for s in seeds:
            print("\n" + "="*80)
            print(f"Model: {m} | Seed: {s}")
            ckpt, hist, test_dl = fit(m, s)

            model = build_model(m, num_classes, pretrained=False).to(DEVICE)
            model.load_state_dict(torch.load(ckpt, map_location=DEVICE))

            y_true, y_pred = predict(model, test_dl)
            cm = confusion_matrix(y_true, y_pred)
            rep = classification_report(y_true, y_pred, output_dict=True, zero_division=0)

            acc = float(rep["accuracy"])
            macro_rec = float(rep["macro avg"]["recall"])
            macro_f1 = float(rep["macro avg"]["f1-score"])
            macro_sens, macro_spec, *_ = macro_sensitivity_specificity(cm)

            rows.append({
                "model": m,
                "seed": s,
                "accuracy": acc,
                "macro_recall": macro_rec,
                "macro_f1": macro_f1,
                "macro_sensitivity": macro_sens,
                "macro_specificity": macro_spec,
                "ckpt": ckpt
            })

    df = pd.DataFrame(rows)
    df.to_csv(os.path.join(cfg.RESULTS_DIR, "benchmark_raw.csv"), index=False)

    metrics = ["accuracy", "macro_recall", "macro_f1", "macro_sensitivity", "macro_specificity"]
    summary = df.groupby("model")[metrics].agg(["mean", "std"]).sort_values(("accuracy","mean"), ascending=False)
    summary.to_csv(os.path.join(cfg.RESULTS_DIR, "benchmark_summary.csv"))
    return df, summary

# Example usage:
ckpt, hist, per_class = run_single("resnet50", seed=2023)
df, summary = run_benchmark(list(cfg.MODELS_TO_RUN), seeds=[2023, 7, 42, 123, 999])
```
